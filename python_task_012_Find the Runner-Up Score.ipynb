# Task-012 : Find the Runner-Up Score!

# Answer :

```python
n = int(input())
arr = list(map(int, input().split()))

if 2 <= n <=10:
    for i in range(n):
        if -100 <= arr[i] <= 100:
            arr.sort()

for i in range(n):
    j = i+1
    if arr[-i] > arr[-j]:
        print(arr[-j])
        break
```
