# Task-010 : Validating and Parsing Email Addresses

# Answer :

```python
import email.utils
import re

n = int(input())

for _ in range(n):
    line = input()

    name, email_addr = email.utils.parseaddr(line)
    pattern = r'^[a-zA-Z][a-zA-Z0-9._-]*@[a-zA-Z]+\.[a-zA-Z]{1,3}$'

    if re.match(pattern, email_addr):
        print(email.utils.formataddr((name, email_addr)))
```
