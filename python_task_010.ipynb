# Task-1 : Say "Hello, World!" With Python

### Description
Here is a simple Python instruction that displays a message:

```python
print("Hello, World!")
```

You can also place the text inside a variable and then print it:

```python
my_string = "Hello, World!"
print(my_string)
```

Both snippets will display the following:

```
Hello, World!
```

### Input Format
No input is required for this task.

### Output Format
The program must print:

```
Hello, World!
```

### Sample Output
```
Hello, World!
```


# Answer :

````python
print("Hello, World!")
````
