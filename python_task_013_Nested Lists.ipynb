# Task-013 : Nested Lists

# Answer :

```python
if __name__ == '__main__':
    n = int(input())
    arr = []
    if 2 <= n <= 5:
        for i in range(n):
            name = input()
            score = float(input())
            arr.append([name, score])

        arr.sort(key=lambda x: x[1])
        for i in range(n):
            j = i+1
            if arr[j][1] > arr[i][1]:
                second_lowest = arr[j][1]
                break
        names_list = []
        for i in arr:
            if second_lowest == i[1]:
                names_list.append(i[0])

        names_list.sort()
        for i in names_list:
            print(i)
```
